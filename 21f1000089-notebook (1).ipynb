```python
X_train = train.drop('Made_Purchase', axis='columns')
Y_train = train['Made_Purchase']

X_test=test
```

```python
import warnings
warnings.filterwarnings('ignore')

train_num=X_train.select_dtypes(include=np.number)
train_obj=X_train.select_dtypes(include=["object"])

test_num=X_test.select_dtypes(include=np.number)
test_obj=X_test.select_dtypes(include=["object"])
```

```python
from sklearn.impute import SimpleImputer
from sklearn.impute import KNNImputer

KNN = KNNImputer(n_neighbors=10, missing_values=np.nan)
SI = SimpleImputer(missing_values=np.nan,strategy="most_frequent")

train_num = pd.DataFrame(KNN.fit_transform(train_num),columns=train_num.columns)
train_obj = pd.DataFrame(SI.fit_transform(train_obj),columns=train_obj.columns)

test_num = pd.DataFrame(KNN.transform(test_num),columns=test_num.columns)
test_obj = pd.DataFrame(SI.transform(test_obj),columns=test_obj.columns)
```

```python
for cols in ['OS', 'SearchEngine','Zone', 'Type of Traffic']:
    train_num[cols] = train_num[cols].astype('object')
    
    
for cols in ['OS', 'SearchEngine','Zone', 'Type of Traffic']:
    test_num[cols] = test_num[cols].astype('object')

train_encoded_cat_col=train_num.iloc[:,10:14]

test_encoded_cat_col=test_num.iloc[:,10:14]
```

```python
train_num = train_num.drop(["OS","SearchEngine",'Zone',"Type of Traffic"],axis='columns')
test_num = test_num.drop(["OS","SearchEngine",'Zone',"Type of Traffic"],axis='columns')
```

```python
from sklearn.preprocessing import MaxAbsScaler
MAS=MaxAbsScaler()

MAS.fit(train_num)
train_num_scaled = pd.DataFrame(MAS.transform(train_num),columns=train_num.columns)
test_num_scaled=pd.DataFrame(MAS.transform(test_num),columns=test_num.columns)


train_num=pd.concat((train_encoded_cat_col,train_num_scaled),axis=1,join="inner")
test_num=pd.concat((test_encoded_cat_col,test_num_scaled),axis=1,join="inner")
```

```python
for cols in ['OS', 'SearchEngine','Zone', 'Type of Traffic']:
    train_num[cols] = train_num[cols].astype('float64')
    
    
for cols in ['OS', 'SearchEngine','Zone', 'Type of Traffic']:
    test_num[cols] = test_num[cols].astype('float64')
```

```python
X_train=pd.concat((train_num,train_obj),axis=1,join="inner")
X_test=pd.concat((test_num,test_obj),axis=1,join="inner")
```

```python
from sklearn.preprocessing import LabelEncoder

import warnings
warnings.filterwarnings('ignore')

LE = LabelEncoder()

object_col = X_train.select_dtypes(include=np.object).columns.tolist()
print(object_col)
# Encode labels in column 'species'.
for col in object_col:
    X_train[col] = LE.fit_transform(X_train[col])
    
```

```python
import warnings
warnings.filterwarnings('ignore')

object_col = X_test.select_dtypes(include=np.object).columns.tolist()
print(object_col)
# Encode labels in column 'species'.
for col in object_col:
    X_test[col] = LE.fit_transform(X_test[col])
```

MODEL

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X_train,Y_train,test_size=0.2,random_state=42)
```

|> Decision tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

dt=DecisionTreeClassifier()

dt.fit(x_train,y_train)
y_pred=dt.predict(x_test)
print(classification_report(y_test,y_pred))
print("Accuracy:",accuracy_score(y_test, y_pred))
```

|> K-neighbours

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score


knn=KNeighborsClassifier()


knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
print(classification_report(y_test,y_pred))
print("Accuracy:",accuracy_score(y_test, y_pred))
```

|>Ranodm Rainforest

```python
from sklearn.ensemble import RandomForestClassifier
RFC=RandomForestClassifier()
RFC.fit(x_train,y_train)

y_pred=RFC.predict(x_test)

from sklearn.metrics import accuracy_score
print(classification_report(y_test,y_pred))
print("Accuracy:",accuracy_score(y_test, y_pred))
```

|> Main Model ===>> Ada booster

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
```

```python
model = AdaBoostClassifier()
```

```python
import warnings
warnings.filterwarnings('ignore')

parameters = {
    'n_estimators': [100,200,300,400],
    'learning_rate' : [0.0001, 0.001, 0.01]
    }
grid_search_ada= GridSearchCV(model, parameters , cv=5 , scoring = 'neg_mean_squared_error', return_train_score=True)

grid_search_ada.fit(X_train,Y_train)

grid_search_ada.best_params_
```
