```python
import numpy as np
import pandas as pd
```

```python
train=pd.read_csv("/kaggle/input/e-commerce-shoppers-behaviour-understanding/train_data_v2.csv")
test=pd.read_csv("/kaggle/input/e-commerce-shoppers-behaviour-understanding/test_data_v2.csv")
```

```python
train.replace('?' , np.nan , inplace= True)
```

```python
# train['Made_Purchase']=train['Made_Purchase'].astype(int)
train.head(5)
```

```python
train.info()
```

```python
train.isnull().sum()
```

```python
X_train = train.drop('Made_Purchase', axis='columns')
Y_train = train['Made_Purchase']
```

```python
X_test=test
```

```python
train_num=X_train.select_dtypes(include=np.number)
train_cat=X_train.select_dtypes(include=["object"])

test_num=X_test.select_dtypes(include=np.number)
test_cat=X_test.select_dtypes(include=["object"])
```

```python
from sklearn.impute import SimpleImputer
si1=SimpleImputer(missing_values=np.nan,strategy="mean")
si2=SimpleImputer(missing_values=np.nan,strategy="most_frequent")

train_num_impute=pd.DataFrame(si1.fit_transform(train_num),columns=train_num.columns)
train_cat_impute=pd.DataFrame(si2.fit_transform(train_cat),columns=train_cat.columns)

test_num_impute=pd.DataFrame(si1.fit_transform(test_num),columns=test_num.columns)
test_cat_impute=pd.DataFrame(si2.fit_transform(test_cat),columns=test_cat.columns)
```

```python
train_cat_impute.isnull().sum()
```

```python
train_num_impute.isnull().sum()
```

```python
test_cat_impute.isnull().sum()
```

```python
test_num_impute.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()

mms.fit(train_num_impute)
num_scaled=pd.DataFrame(mms.transform(train_num_impute),columns=train_num_impute.columns)
```

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()

mms.fit(test_num_impute)
num_scaled_test=pd.DataFrame(mms.transform(test_num_impute),columns=test_num_impute.columns)
```

```python
num_scaled
```

```python
X_train=pd.concat((num_scaled,train_cat_impute),axis=1,join="inner")
```

```python
X_test=pd.concat((num_scaled_test,test_cat_impute),axis=1,join="inner")
```

```python
from sklearn.preprocessing import LabelEncoder
  
label_encoder = LabelEncoder()

object_col = X_train.select_dtypes(include=np.object).columns.tolist()
print(object_col)
# Encode labels in column 'species'.
for col in object_col:
    X_train[col] = label_encoder.fit_transform(X_train[col])
```

```python
from sklearn.preprocessing import LabelEncoder
  
label_encoder = LabelEncoder()

object_col = X_test.select_dtypes(include=np.object).columns.tolist()
print(object_col)
# Encode labels in column 'species'.
for col in object_col:
    X_test[col] = label_encoder.fit_transform(X_test[col])
```
